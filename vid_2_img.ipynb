```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load video
# change it to your video location
vid_location = r"C:\Users\ASUS\Desktop\motion tracking example data\4d_cell_ring\1.mp4"

cap = cv2.VideoCapture(vid_location)
n_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Get video width (assuming success)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))

# Get video height (assuming success)
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# store all frames
vidframes = np.zeros((height, width, 3, n_frames))

# read all frames 
for i in range(n_frames):
 ret, frame = cap.read()
 #print(i)
    
 if not ret:
  print("Can't receive frame (stream end?). Exiting ...")
  break

 vidframes[:,:,:,i] = frame

cap.release()
```

```python
# check the histogram of all frames
vidframes_flat = vidframes.ravel()

plt.hist(vidframes_flat)
plt.show()
```

```python
# hand pick new min and max of the window
new_min = 30
new_max = 80

# transform
new_vidframes = (vidframes-new_min)/(new_max-new_min)*255
# get rid of numbers exceeding 0 and 255
index1 = np.where(new_vidframes>255)
new_vidframes[index1] = 255
index2 = np.where(new_vidframes<0)
new_vidframes[index2] = 0

# change data type to integer for making images
frames_int = new_vidframes.astype(int)
```

```python
# check the frame after changing the window
plt.imshow(frames_int[:,:,:,0])
plt.show()
```

```python
# make images
# change it to your image folder
img_location = r'C:\desktop-everything\dataAnalysis\rnd_script\img_export'

for i in range(n_frames):
 #img_name1 = f"f{i:0>6}" + ".jpeg"
 img_name1 = f"f{i:0>6}" + ".tiff"
 img_name = os.path.join(img_location,img_name1)
 #print(img_name)
 cv2.imwrite(img_name,frames_int[:,:,:,i])
```
